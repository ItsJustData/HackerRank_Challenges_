```
# Enter your code here. Read input from STDIN. Print output to STDOUT

import sys 
import numpy as np
from sklearn.preprocessing import PolynomialFeatures 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

F_N = sys.stdin.readline().strip().split()

num_features = int(F_N[0])
num_rows = int(F_N[1])

degree = 3

X = []
y = []

for i in range(num_rows):
    line_train = sys.stdin.readline().strip().split()
    X.append(line_train[0:num_features])
    y.append(line_train[num_features])

X = np.asarray(X)
y = np.asarray(y)

X.astype(float)
y.astype(float)

X_train, X_test, y_train, y_test = train_test_split(X, y, 
test_size=0.1,
random_state=10)

poly = PolynomialFeatures(degree=degree)
X_train_poly = poly.fit_transform(X_train)

model = LinearRegression()
model.fit(X_train_poly, y_train)


N = sys.stdin.readline().strip().split()

val_rows = int(N[0])

X_val = []

for i in range(val_rows):
    line_val = sys.stdin.readline().strip().split()
    X_val.append(line_val)

X_val = np.asarray(X_val)
X_val.astype(float)

X_val_poly = poly.fit_transform(X_val)

y_hat = model.predict(X_val_poly)

for prediction in y_hat:
    print(round(prediction,2))

  

```
